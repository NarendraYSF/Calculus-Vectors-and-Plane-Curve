```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the parametric equations for sec(t) and tan(t)
def sec_t(t):
    return 1 / np.cos(t)

def tan_t(t):
    return np.tan(t)

# Define the range of t, avoiding points where sec(t) and tan(t) become undefined
t_values = np.linspace(0, 2 * np.pi, 1000)
t_values = t_values[np.abs(np.cos(t_values)) > 0.1]  # Avoid values where sec(t) is undefined

# Calculate x and y components of the vector r(t)
x_values = sec_t(t_values)
y_values = tan_t(t_values)

# Plotting the curve
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$\mathbf{r}(t) = \sec(t) \hat{i} + \tan(t) \hat{j}$', color='b')
plt.title('Plane Curve of r(t) = sec(t) i + tan(t) j')
plt.xlabel('x = sec(t)')
plt.ylabel('y = tan(t)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
