```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Define the parametric equations for r(t)
def r_t(t):
    sec_t = 1 / np.cos(t)
    tan_t = np.tan(t)
    return sec_t, tan_t

# Generate values for t from 0 to pi/2 (we'll avoid pi/2 where sec(t) and tan(t) go to infinity)
t_values = np.linspace(0, np.pi / 2 - 0.1, 400)
```

```python
# Compute r(t) values for the range of t
x_values, y_values = r_t(t_values)

# Plot the curve
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$\mathbf{r}(t) = \sec(t) \hat{i} + \tan(t) \hat{j}$', color='b')
plt.scatter([np.sqrt(2)], [1], color='r', label=r'$t = \frac{\pi}{4}$', zorder=5)

# Adding labels and title
plt.title('Sketch of the Curve for r(t) = sec(t) i + tan(t) j')
plt.xlabel('x (sec(t))')
plt.ylabel('y (tan(t))')

# Show grid, legend, and the plot
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend(loc='best')
plt.show()
```

```python
# Define the functions for r(t) and r'(t)
def r_t(t):
    return np.array([1/np.cos(t), np.tan(t)])

def r_prime_t(t):
    return np.array([1/np.cos(t) * np.tan(t), 1/np.cos(t)**2])

# Value of t = pi/4
t = np.pi / 4

# Compute r(t) and r'(t) at t = pi/4
r_at_t = r_t(t)
r_prime_at_t = r_prime_t(t)

# Plotting the vectors
plt.figure(figsize=(8, 6))

# Plot r(t) (Position Vector)
plt.quiver(0, 0, r_at_t[0], r_at_t[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\mathbf{r}(t)$')

# Plot r'(t) (Derivative Vector)
plt.quiver(0, 0, r_prime_at_t[0], r_prime_at_t[1], angles='xy', scale_units='xy', scale=1, color='r', label=r"$\mathbf{r}'(t)$")

# Labels and plot settings
plt.xlim(-2, 3)
plt.ylim(-2, 3)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title(r"Vectors $\mathbf{r}(t)$ and $\mathbf{r}'(t)$ at $t = \frac{\pi}{4}$")
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# Show plot
plt.show()
```

```python
# Define the components of the vector r(t)
r_i = t               # i-component
r_j = 2 * sp.sin(t)   # j-component
r_k = 3 * sp.cos(t)   # k-component

# Compute the derivative of r(t) (tangent vector r'(t))
dr_i_dt = sp.diff(r_i, t)  # Derivative of i-component
dr_j_dt = sp.diff(r_j, t)  # Derivative of j-component
dr_k_dt = sp.diff(r_k, t)  # Derivative of k-component
```

```python
# Construct the derivative vector r'(t)
i, j, k = sp.symbols('i j k')  # Define unit vectors i, j, k
r_prime = dr_i_dt * i + dr_j_dt * j + dr_k_dt * k  # Final vector derivative

# Compute the magnitude of r'(t)
magnitude_r_prime = sp.sqrt(dr_i_dt**2 + dr_j_dt**2 + dr_k_dt**2)

# Unit tangent vector T(t) = r'(t) / |r'(t)|
T_t = r_prime / magnitude_r_prime

# Evaluate the unit tangent vector at t = pi/6
T_at_pi_6 = T_t.subs(t, sp.pi/6)

# Display the result
T_at_pi_6.simplify()
```

```python
# Define the components of the vector function
i_component = sp.cos(2*t)  # i-component
j_component = sp.sin(2*t)  # j-component
k_component = t * sp.sin(t)  # k-component

# Perform the integrals for each component
i_integral = sp.integrate(i_component, (t, 0, sp.pi/4))
j_integral = sp.integrate(j_component, (t, 0, sp.pi/4))
k_integral = sp.integrate(k_component, (t, 0, sp.pi/4))

# Redefine the unit vectors as regular symbols
i, j, k = sp.symbols('i j k')

# Construct the final result as a vector using regular symbols for i, j, k
result_vector = i_integral * i + j_integral * j + k_integral * k

# Display the result
result_vector.simplify()
```

```python
# Define time variable as a symbol
t = sp.symbols('t')

# Define position functions (x(t), y(t), z(t))
x = 2 * sp.sin(3 * t)
y = 2 * sp.cos(3 * t)
z = 8 * t

# Define the position vector
position_vector = sp.Matrix([x, y, z])

# Velocity is the first derivative of the position
velocity_vector = position_vector.diff(t)

# Acceleration is the first derivative of the velocity
acceleration_vector = velocity_vector.diff(t)

# Define the magnitude of the velocity vector
velocity_magnitude = sp.sqrt(velocity_vector[0]**2 + velocity_vector[1]**2 + velocity_vector[2]**2)

# Output the symbolic expressions for velocity and acceleration
print("Velocity:")
print(f"v(t) = {velocity_vector[0]} i^ - {velocity_vector[1]} j^ + {velocity_vector[2]} k^")

print("\nAcceleration:")
print(f"a(t) = {acceleration_vector[0]} i^ - {acceleration_vector[1]} j^ + {acceleration_vector[2]} k^")

# Now, compute the velocity and acceleration at t = 90 degrees (t = pi/2)
t_value = np.pi / 2
velocity_at_90 = velocity_vector.subs(t, t_value)
acceleration_at_90 = acceleration_vector.subs(t, t_value)

# Convert symbolic results to float for cleaner output and round small values to 0
velocity_at_90 = [round(val, 6) if abs(val) > 1e-6 else 0 for val in velocity_at_90]
acceleration_at_90 = [round(val, 6) if abs(val) > 1e-6 else 0 for val in acceleration_at_90]

# Compute the magnitude of the velocity at t = 90 degrees
velocity_magnitude_90 = velocity_magnitude.subs(t, t_value)
velocity_magnitude_90 = round(velocity_magnitude_90, 6)

# Output the results at t = 90 degrees
print("\nVelocity at t = 90 degrees (t = pi/2):")
print(f"v(90 degrees) = {velocity_at_90[0]} i^ - {velocity_at_90[1]} j^ + {velocity_at_90[2]} k^")

print("\nAcceleration at t = 90 degrees (t = pi/2):")
print(f"a(90 degrees) = {acceleration_at_90[0]} i^ - {acceleration_at_90[1]} j^ + {acceleration_at_90[2]} k^")

print(f"\nMagnitude of velocity at t = 90 degrees (t = pi/2): {velocity_magnitude_90}")
```
